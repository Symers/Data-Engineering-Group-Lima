```python
import pandas as pd

datafile1 = "financial_compliance_source.csv"
datafile2 = "financial_compliance_target.csv"
targetTable = pd.read_csv(datafile2)
inputTable = pd.read_csv(datafile1)
```

```python
#Uppercase and Fillna

inputTable.columns = [column.lower() for column in inputTable.columns]
inputTable = inputTable.fillna('NA')
inputTable
```

```python
#AI Mapping and Overall Score

inputTable["overall_score"] = None
for x in range(len(inputTable)):
    word = str(inputTable.loc[x, "ai_used_for_auditing"])
    audit = inputTable.loc[x, "audit_effectiveness_score"]
    client = inputTable.loc[x, "client_satisfaction_score"]
    
    if word.lower() == "yes":
        inputTable.loc[x, "ai_used_for_auditing"] = 1
    elif str(word).lower() == "no":
        inputTable.loc[x, "ai_used_for_auditing"] = 0

    #If either score is empty, assign NA and go to next row
    if str(audit) == "NA" or str(client) == "NA":
        inputTable.loc[x,"overall_score"] = "NA"
        continue

    auditScore = float(audit)
    clientScore = float(client)
    overallScore = (auditScore + clientScore) / 2

    inputTable.loc[x,"overall_score"] = overallScore

inputTable
```
